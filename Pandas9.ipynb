```python
import pandas as pd

path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'

df = pd.read_excel(path)
df.head()
```

```python
df['Spending Percentage'] = (df['Total Spent'] / df['Total Spent'].sum()) * 100
print("\n1. Spending Percentage for each customer:\n", df[['Customer ID', 'Spending Percentage']])
```

```python
df['Normalized Return Rate'] = (df['Return Rate (%)'] - df['Return Rate (%)'].min()) / (df['Return Rate (%)'].max() - df['Return Rate (%)'].min())
print("\n2. Normalized Return Rate:\n", df[['Customer ID', 'Normalized Return Rate']])
```

```python
df['Age Group'] = pd.cut(df['Age'], bins=[0, 25, 40, 60], labels=['Young', 'Middle-aged', 'Older'])
print("\n3. Age Groups:\n", df[['Customer ID', 'Age', 'Age Group']])
```

```python
df['High Discount'] = df['Discount Availed'].apply(lambda x: 'Yes' if x > 50 else 'No')
print("\n4. High Discount Flag:\n", df[['Customer ID', 'Discount Availed', 'High Discount']])
```

```python
df['Cumulative Spending'] = df['Total Spent'].cumsum()
print("\n5. Cumulative Spending:\n", df[['Customer ID', 'Total Spent', 'Cumulative Spending']])
```

```python
correlation = df['Age'].corr(df['Total Spent'])
print("\n6. Correlation between Age and Total Spent:\n", correlation)
```

```python
df['Purchased Category'].fillna('Unknown', inplace=True)
print("\n7. Filled Missing 'Purchased Category':\n", df[['Customer ID', 'Purchased Category']])
```

```python
high_spenders = df[df['Total Spent'] > 500]
print("\n8. High Spenders (Total Spent > 500):\n", high_spenders)
```

```python
region_avg_discount = df.groupby('Region')['Discount Availed'].mean()
print("\n9. Average Discount Availed by Region:\n", region_avg_discount)
```

```python
sorted_df = df.sort_values(by='Total Spent', ascending=False)
print("\n10. Data Sorted by Total Spent (Descending):\n", sorted_df)
```
