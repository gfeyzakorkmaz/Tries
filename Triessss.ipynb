```python
import pandas as pd
from google.colab import files
from textblob import TextBlob
import matplotlib.pyplot as plt


path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'
df = pd.read_excel(path)


df['Sentiment'] = df['Comments'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
print(df[['Comments', 'Sentiment']].head())


average_sentiment = df.groupby('Region')['Sentiment'].mean()
plt.figure(figsize=(8, 6))
average_sentiment.plot(kind='bar', title='Average Sentiment by Region')
plt.ylabel('Average Sentiment')
plt.savefig('average_sentiment_by_region.png')  # Grafiği PNG formatında kaydet
plt.show()


age_bins = pd.cut(df['Age'], bins=[0, 20, 30, 40, 50, 60, 70], labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61-70'])
age_spending = df.groupby(age_bins)['Total Spent'].mean()
plt.figure(figsize=(8, 6))
age_spending.plot(kind='bar', title='Spending by Age', xlabel='Age Group', ylabel='Average Spending')
plt.savefig('spending_by_age.png')  # Grafiği PNG formatında kaydet
plt.show()


df.to_excel('data_with_sentiment.xlsx', index=False)


files.download('data_with_sentiment.xlsx')
files.download('average_sentiment_by_region.png')
files.download('spending_by_age.png')
```
