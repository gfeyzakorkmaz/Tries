```python
import pandas as pd

path = '/content/drive/MyDrive/Dataset/Superstore 2018 2021.csv'

df = pd.read_csv(path)
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

weekly_trend = df.groupby([pd.Grouper(key='Order Date', freq='W'), 'Discount'])['Sales'].sum().reset_index()

print(weekly_trend.head())
```

```python
january_spending = df[df['Order Date'].dt.month == 1]

january_summary = january_spending.groupby('Discount')['Sales'].sum().reset_index()

print(january_summary.head())
```

```python
february_spending = df[df['Order Date'].dt.month == 2]

february_summary = february_spending.groupby('Discount')['Sales'].sum().reset_index()

print(february_summary.head())
```

```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_spending =df.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales'].sum()

monthly_spending_growth = monthly_spending.pct_change().reset_index(name='growth_rate')

print(monthly_spending_growth.head())
```
