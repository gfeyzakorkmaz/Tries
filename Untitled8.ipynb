```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Dataset/Orders.xlsx'

df = pd.read_excel(path)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
df.groupby('Profit')['Sales'].sum().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Total Sales by Profit")
plt.xlabel("Profit")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index.astype(str), monthly_sales.values, marker='o', linestyle='-', color='b', label='Monthly Sales')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Profit'], bins=30, kde=True, color='green')
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Order Quantity")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Discount'], y=df['Sales'], alpha=0.6, color='red')
plt.title("Discount vs Sales")
plt.xlabel("Discount")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()
```

```python
state_profits = df.groupby('State')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(state_profits.index, state_profits.values, color='purple')
plt.title("Total Profit by State")
plt.xlabel("State")
plt.ylabel("Total Profit")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Ship Mode'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'gold'])
plt.title("Order Distribution by Ship Mode")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x=df['Sales'], y=df['Profit'], scatter_kws={'alpha':0.5}, line_kws={"color": "red"})
plt.title("Sales vs. Profit Relationship")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

```python
top_products = df.groupby('Ship Mode')['Profit'].sum().nlargest(10)

plt.figure(figsize=(12, 6))
top_products.plot(kind='barh', color='orange', edgecolor='black')
plt.title("Top 10 Most Profitable Products")
plt.xlabel("Total Profit")
plt.ylabel("Ship Mode")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Order Priority'], y=df['Profit'], palette="coolwarm")
plt.title("Profit Distribution by Customer Segment")
plt.xlabel("Order Priority")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```
