```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Dataset/Orders.xlsx'

df = pd.read_excel(path)
df.head()
```

```python
total_revenue = df.groupby('Category')['Sales'].sum().reset_index()
print(total_revenue)
```

```python
most_profitable_product = df.groupby('SubRegion').apply(lambda x: x.loc[x['Profit'].idxmax()])
print(most_profitable_product[['SubRegion', 'Customer Name', 'Profit']])
```

```python
avg_shipping_cost = df.groupby('Customer')['Shipping Cost'].mean().reset_index()
print(avg_shipping_cost)
```

```python
df['Profit'] = (df['Profit'] / df['Sales']) * 100
print(df[['Customer Name', 'Profit']].head())
```

```python
shipping_profit_correlation = df['Shipping Cost'].corr(df['Profit'])
print(f"Correlation between Shipping Cost and Profit: {shipping_profit_correlation}")
```

```python
top_10_customers = df.groupby('Customer Name')['Sales'].sum().nlargest(10).reset_index()
print(top_10_customers)
```

```python
low_performers = df[(df['Sales'] < 500) & (df['Profit'] < 50)]
print(low_performers[['Customer', 'Sales', 'Profit']])
```

```python
df['Profitability'] = df['Profit'].apply(lambda x: 'High Profit' if x > df['Profit'].mean() else 'Low Profit')
print(df[['Customer', 'Profit', 'Profitability']].head())
```

```python
most_expensive_shipping = df.groupby('SubRegion')['Shipping Cost'].mean().idxmax()
print(f"The most expensive subregion to ship: {most_expensive_shipping}")
```

```python
sales_profit_pivot = df.pivot_table(values=['Sales', 'Profit'], index='State', columns='SubRegion', aggfunc='sum')
print(sales_profit_pivot)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

df['Oder Date'] = np.arange(len(df))
X = df[['Order Date']]
y = df['Sales']

model = LinearRegression().fit(X, y)
future_time = np.array([[len(df) + i] for i in range(1, 11)])
future_sales = model.predict(future_time)

print(f"Predicted sales for the next 10 periods: {future_sales}")
```
