```python
import pandas as pd

path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'

df = pd.read_excel(path)
df.head()

pivoted = df.pivot_table(
    values='Total Spent',
    index='Region',
    columns='Purchased Category',
    aggfunc='sum',
    fill_value=0
)

unpivoted = pivoted.reset_index().melt(
    id_vars='Region',
    var_name='Purchased Category',
    value_name='Total Spent'
)

print("\nUnpivoted Data:")
print(unpivoted)
```
