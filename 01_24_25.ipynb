```python
import pandas as pd

path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'

df = pd.read_excel(path)
df.head()
```

```python
from sklearn.cluster import KMeans

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['scaled_spent', 'scaled_return']] = scaler.fit_transform(df[['Total Spent', 'Return Rate (%)']])

kmeans = KMeans(n_clusters=4, random_state=42)
df['income_level'] = kmeans.fit_predict(df[['scaled_spent', 'scaled_return']])

df[['Customer ID', 'Total Spent', 'Return Rate (%)', 'Income Level']].to_csv('customer_segments_tableau.csv', index=False)

segment_summary = df.groupby('income_level').agg({
    'Total Spent': 'mean',
    'Return Rate (%)': 'mean'
}).reset_index()
print(segment_summary)

from google.colab import files
files.download('customer_segments_tableau.csv')
```

```python
df['roi'] = (df['Total Spent'] - df['Total Spent Diff']) / df['Total Spent Diff']
campaign_analysis = df.groupby('Campaign').agg({
    'Total Spent': 'sum',
    'Total Spent Diff': 'sum',
    'roi': 'mean'
}).reset_index()

campaign_analysis.to_csv('campaign_effectiveness_tableau.csv', index=False)
from google.colab import files
files.download('campaign_effectiveness_tableau.csv')
```
