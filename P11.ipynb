```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'
```

```python
df = pd.read_excel(path)
df.head(3)
```

```python
correlation = df['Total Spent'].corr(df['Return Rate (%)'])
print("Correlation between rates and spending:", correlation)
```

```python
df['roi'] = (df['Total Spent'] - df['Return Rate (%)']) / df['Return Rate (%)']
campaign_roi = df.groupby('Campaign')['roi'].mean().reset_index()
campaign_roi
```

```python
df['churn_risk'] = pd.cut(df['Return Rate (%)'], bins=[0, 0.25, 0.75, 1], labels=['Low', 'Medium', 'High'])
df[['Return Rate (%)', 'churn_risk']].head()
```

```python
spending_by_region_age = df.groupby(['Region', 'Age'])['Total Spent'].sum().unstack()
spending_by_region_age
```

```python
df['conversion_rate'] = (df['Converted'] / df['Website Visit Frequency']).fillna(0)
df[['Website Visit Frequency', 'Converted', 'conversion_rate']].head()
```

```python
pivot_table = df.pivot_table(values=['Total Spent', 'Return Rate (%)'], index='Campaign', columns='Region', aggfunc='mean')
pivot_table
```

```python
df['spending_segment'] = pd.cut(df['Total Spent'], bins=[0, 500, 1000, 5000], labels=['Low', 'Medium', 'High'])
df[['Total Spent', 'spending_segment']].head()
```

```python
summary_for_tableau = df.groupby(['Campaign', 'Region']).agg({
    'Total Spent': 'sum',
    'Return Rate (%)': 'mean',
    'Spending Rank': 'mean'
}).reset_index()

summary_for_tableau.to_csv('summary_for_tableau.csv', index=False)

print(summary_for_tableau)
```
