```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialize Faker and set random seed for reproducibility
fake = Faker()
np.random.seed(42)
random.seed(42)

# Generate synthetic dataset
num_records = 1000
categories = ['Electronics', 'Fashion', 'Home', 'Sports', 'Toys']
sub_categories = {
    'Electronics': ['Phones', 'Laptops', 'Accessories'],
    'Fashion': ['Clothing', 'Footwear', 'Accessories'],
    'Home': ['Furniture', 'Decor', 'Kitchen'],
    'Sports': ['Outdoor', 'Gym', 'Team Sports'],
    'Toys': ['Educational', 'Action Figures', 'Board Games']
}
payment_methods = ['Credit Card', 'PayPal', 'Bank Transfer']
delivery_statuses = ['Delivered', 'Pending', 'Canceled']
churn_risk_levels = ['High', 'Medium', 'Low']

data = []

for _ in range(num_records):
    category = random.choice(categories)
    sub_category = random.choice(sub_categories[category])
    quantity = random.randint(1, 10)
    unit_price = round(random.uniform(10, 1000), 2)
    discount = round(random.uniform(0, 0.3), 2)
    total_sales = round(quantity * unit_price * (1 - discount), 2)
    profit = round(total_sales * random.uniform(0.05, 0.3), 2)
    feedback_score = random.randint(1, 5)
    churn_risk = random.choices(churn_risk_levels, weights=[1, 2, 3], k=1)[0] if feedback_score < 3 else 'Low'

    data.append({
        'Order ID': fake.uuid4(),
        'Customer ID': fake.uuid4(),
        'Customer Name': fake.name(),
        'Gender': random.choice(['Male', 'Female', 'Other']),
        'Age': random.randint(18, 65),
        'Country': fake.country(),
        'City': fake.city(),
        'Order Date': fake.date_between(start_date='-4y', end_date='today'),
        'Product ID': fake.uuid4(),
        'Product Name': fake.word().capitalize(),
        'Category': category,
        'Sub-Category': sub_category,
        'Quantity': quantity,
        'Unit Price': unit_price,
        'Total Sales': total_sales,
        'Discount (%)': discount * 100,
        'Profit': profit,
        'Payment Method': random.choice(payment_methods),
        'Delivery Status': random.choice(delivery_statuses),
        'Customer Feedback Score': feedback_score,
        'Churn Risk': churn_risk
    })

# Create DataFrame
df = pd.DataFrame(data)

# Save to Excel
file_path = '/mnt/data/synthetic_ecommerce_data.xlsx'
df.to_excel(file_path, index=False)

file_path
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

file_path = '/content/drive/MyDrive/Dataset/Orders.xlsx'
df = pd.read_excel(file_path)


df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

df.fillna(0, inplace=True)


df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Month'].astype(str), monthly_sales['Sales'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid()
plt.xticks(rotation=45)
plt.show()


region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=region_sales.values, y=region_sales.index, palette='viridis')
plt.title('Sales by Region')
plt.xlabel('Sales')
plt.ylabel('Region')
plt.show()


category_sales = df.groupby(['Category', 'SubRegion'])['Sales'].sum().unstack()

plt.figure(figsize=(12, 6))
sns.heatmap(category_sales, annot=True, fmt='.0f', cmap='coolwarm')
plt.title('Sales by Category and Sub-Category')
plt.show()


if 'Campaign' in df.columns:
    campaign_sales = df.groupby('Campaign')['Sales'].sum().sort_values(ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=campaign_sales.values, y=campaign_sales.index, palette='magma')
    plt.title('Campaign-wise Sales Performance')
    plt.xlabel('Sales')
    plt.ylabel('Campaign')
    plt.show()


if 'Converted' in df.columns:
    conversion_rate = df.groupby('Campaign')['Converted'].mean() * 100

    plt.figure(figsize=(10, 6))
    sns.barplot(x=conversion_rate.values, y=conversion_rate.index, palette='rocket')
    plt.title('Conversion Rate by Campaign')
    plt.xlabel('Conversion Rate (%)')
    plt.ylabel('Campaign')
    plt.show()


segment_profit = df.groupby('Order Priority')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=segment_profit.values, y=segment_profit.index, palette='cubehelix')
plt.title('Profit by Segment')
plt.xlabel('Profit')
plt.ylabel('Segment')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Region', alpha=0.7)
plt.title('Sales vs Profit Correlation')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid()
plt.show()


summary = df.groupby(['Region', 'Category']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
summary.to_csv('/content/drive/MyDrive/Dataset/Orders.xlsx', index=False)
```
