```python
import pandas as pd
from google.colab import files

path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'

df = pd.read_excel(path)

income_level_spending = df.groupby('Income Level')['Total Spent'].mean()
print("\n16. Mean Total Spent by Income Level:\n", income_level_spending)

customer_comments = df[df['Comments'].str.contains('customer', case=False, na=False)]
print("\n17. Rows where Comments contain 'customer':\n", customer_comments)

df['Frequent Visitor'] = df['Website Visit Frequency'] > 5
print("\n18. Frequent Visitor Flag:\n", df[['Customer ID', 'Website Visit Frequency', 'Frequent Visitor']])

df['Spending Per Visit'] = df['Total Spent'] / (df['Website Visit Frequency'] + 1)
print("\n19. Spending Per Visit:\n", df[['Customer ID', 'Total Spent', 'Website Visit Frequency', 'Spending Per Visit']])

most_common_campaign = df['Campaign'].mode()[0]
print("\n20. Most Common Campaign:\n", most_common_campaign)


df.to_csv('marketing_data.csv', index=False)
files.download('marketing_data.csv')
```

```python
import pandas as pd
from google.colab import files

path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'

df = pd.read_excel(path)

df['Discount to Spending Ratio'] = df['Discount Availed'] / df['Total Spent']
print("\n21. Discount to Spending Ratio:\n", df[['Customer ID', 'Total Spent', 'Discount Availed', 'Discount to Spending Ratio']])

region_gender_spending = df.groupby(['Region', 'Gender'])['Total Spent'].sum()
print("\n22. Total Spent by Region and Gender:\n", region_gender_spending)

df['Region Discount Percentage'] = df.groupby('Region')['Discount Availed'].transform(lambda x: (x / x.sum()) * 100)
print("\n23. Region Discount Percentage:\n", df[['Customer ID', 'Region', 'Discount Availed', 'Region Discount Percentage']])

df['Spending Z-Score'] = (df['Total Spent'] - df['Total Spent'].mean()) / df['Total Spent'].std()
outliers = df[df['Spending Z-Score'].abs() > 2]
print("\n24. Outliers in Total Spent:\n", outliers[['Customer ID', 'Total Spent', 'Spending Z-Score']])

threshold = df['Total Spent'].quantile(0.9)
df['Top 10% Spender'] = df['Total Spent'] >= threshold
print("\n25. Top 10% Spender Flag:\n", df[['Customer ID', 'Total Spent', 'Top 10% Spender']])

df.to_csv('marketing_data.csv', index=False)
files.download('marketing_data.csv')
```

```python
import pandas as pd
from google.colab import files

path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'

df = pd.read_excel(path)

age_group_spending = df.groupby('Age')['Total Spent'].mean()
print("\n26. Average Spending by Age Group:\n", age_group_spending)

import datetime
df['Joined'] = pd.to_datetime(['2021-01-01', '2020-05-15', '2021-07-10', '2019-11-05', '2020-08-20'])
df['Days Since Joining'] = (datetime.datetime.now() - df['Joined']).dt.days
print("\n27. Days Since Joining:\n", df[['Customer ID', 'Joined', 'Days Since Joining']])

above_average_cart = df[df['Average Cart Value'] > df['Average Cart Value'].mean()]
print("\n28. Customers with Above-Average Cart Value:\n", above_average_cart)

df['Spending Category'] = pd.cut(df['Total Spent'], bins=[0, 300, 700, df['Total Spent'].max()], labels=['Low', 'Medium', 'High'])
print("\n29. Spending Categories:\n", df[['Customer ID', 'Total Spent', 'Spending Category']])

top_5_spending = df.nlargest(5, 'Total Spent')[['Total Spent', 'Discount Availed']].sum()
print("\n30. Sum of Total Spent and Discount Availed for Top 5 Spenders:\n", top_5_spending)

df.to_csv('marketing_data.csv', index=False)
files.download('marketing.csv')
```
