```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/TechCompany_Extended_CaseStudy.xlsx'

df = pd.read_excel(path)
df.head()
```

```python
weighted_conversion = df.groupby(['Year', 'Total Sales ($)']).apply(
    lambda x: (x['Desktops'] * x['Accessories']).sum() / x['Accessories'].sum()
).reset_index(name='Weighted Conversion Rate')
weighted_conversion
```

```python
df['Segment'] = pd.cut(
    df['Total Sales ($)'],
    bins=[0, 1000, 2000, 3000],
    labels=['Low Sales', 'Mid Sales', 'High Sales']
)

df['Loyalty Level'] = pd.cut(
    df['Total Sales ($)'],
    bins=[0, 1000, 2000, 3000],
)

segmentation = df.groupby(['Segment', 'Loyalty Level']).size().unstack(fill_value=0)
segmentation
```

```python
spend_mean = df['Total Sales ($)'].mean()
spend_std = df['Total Sales ($)'].std()

df['Anomaly'] = (df['Total Sales ($)'] > (spend_mean + 3 * spend_std)) | \
                  (df['Total Sales ($)'] < (spend_mean - 3 * spend_std))
df[['Year', 'Total Sales ($)', 'Anomaly']]
```

```python
retention_rate = df.groupby('Year').apply(
    lambda x: x[x['Total Sales ($)'] == 1].shape[0] / x.shape[0]
).reset_index(name='Retention Rate')
retention_rate
```
