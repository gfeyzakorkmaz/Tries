```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df.head()
```

```python
df['total_spent'] = df['total_spent'].fillna(df['total_spent'].mean())
```

```python
high_spenders = df[df['total_spent'] > 100]
```

```python
df['age_group'] = pd.cut(df['age'], bins=[18, 30, 45, 60, 100], labels=['18-30', '31-45', '46-60', '60+'])
df[['age', 'age_group']].head()
```

```python
spending_by_campaign = df.groupby('campaign')['total_spent'].sum().reset_index()
spending_by_campaign
```

```python
return_rate_category = df.groupby('purchased_category')['return_rate_(%)'].mean().reset_index()
return_rate_category
```

```python
df['loyal_customer'] = df['customer_tenure_quarter']
df[['customer_tenure_quarter', 'loyal_customer']].head()
```
